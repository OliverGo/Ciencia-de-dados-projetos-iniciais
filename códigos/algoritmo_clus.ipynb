```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.cluster import KMeans


np.random.seed(0)

n_samples = 500

 

data = {

    'valor_medio_gasto': np.random.randn(n_samples) * 20 + 200,

    'frequencia_compras': np.random.randn(n_samples) * 5 + 15,

    'categorias_produtos': np.random.randn(n_samples) * 2 + 8

}

 

df = pd.DataFrame(data)



 

data = {

    'valor_medio_gasto': np.random.randn(n_samples) * 20 + 200,

    'frequencia_compras': np.random.randn(n_samples) * 5 + 15,

    'categorias_produtos': np.random.randn(n_samples) * 2 + 8

}

 

df = pd.DataFrame(data)

print(df.head())


X = df[['valor_medio_gasto', 'frequencia_compras', 'categorias_produtos']].values 

KMeans = KMeans(n_clusters=3, random_state=0)
KMeans.fit(X)

df['cluster'] = KMeans.predict(X)


plt.scatter(df['valor_medio_gasto'], df['frequencia_compras'], c=df['cluster'], s= 10)

centers = KMeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha=0.5)
plt.title('Segmentao de Clientes')
plt.xlabel('Valor mdio gasto')
plt.ylabel('Frequncia de compras')
plt.show()
```


Por que esse algoritmo  adequado para o problema apresentado?

- Foi utilizado Kmeans para a resoluo do problema, pois ele  um algoritmo de aprendizado no supervisionado que ajuda a prever 
rotulos, nesse caso perfis de clientes 

Quais so as vantagens e possveis limitaes do algoritmo escolhido no contexto deste problema?

- O algoritmo se comporta muito bem com grandes quantidades de dados sendo rapido para definir grupos. Entretanto o Kmeans  sensivel a outliers,
podendo afetar o resultado final, como definir grupos errados 

Como os clusters formados podem ser interpretados para gerar insights para a loja on-line?

- Ajudando a entender os perfis dos clientes. Por exemplo entender que aquele cliente frequentimente gasta um grande valor naquela loja.
